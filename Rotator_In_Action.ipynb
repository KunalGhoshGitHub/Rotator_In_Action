```python
Shape = [15,15]
```

```python
A = np.random.rand(Shape[0],Shape[1])
```

```python
A_copy = A.copy()
Q = np.eye(Shape[0])
Q_temp = np.eye(Shape[0])
Image_Files = []
ctr = 0
for j in range(Shape[1]):
    for k in  range(j+1,Shape[0]):
        # d = np.sqrt((A_copy[j,j]**2)+(A_copy[k,j]**2))
        # theta = np.arctan(A_copy[k,j]/A_copy[j,j])
        theta = np.arctan2(A_copy[k,j],A_copy[j,j])
        # A_copy[j,j] = d
        # A_copy[k,j] = 0
        Q_temp = np.eye(Shape[0])
        Q_temp[j,j] = np.cos(theta)
        Q_temp[k,k] = np.cos(theta)
        Q_temp[j,k] = -np.sin(theta)
        Q_temp[k,j] = np.sin(theta)
        Q = Q@Q_temp
        A_copy = Q_temp.T@A_copy
        # print(f"A: \n{A_copy}\n")
        # print(f"Q_temp: \n{Q_temp}\n")
        # print(f"j: {j}, k: {k}, d: {d}\n")
        # sns.heatmap(A_copy,cmap="viridis",annot=True,vmin=-1.5,vmax=1.5)
        # sns.heatmap(A_copy,cmap="viridis",vmin=Min,vmax=Max)
        # plt.savefig(f"Plot_{ctr}.png")
        # Image_Files.append(f"Plot_{ctr}.png")
        ctr = ctr + 1
        # plt.show()
```

```python
Max = A_copy.max()
```

```python
Min = A_copy.min()
```

```python
A_copy = A.copy()
Q = np.eye(Shape[0])
Q_temp = np.eye(Shape[0])
Image_Files = []
ctr = 0
for j in range(Shape[1]):
    for k in  range(j+1,Shape[0]):
        # d = np.sqrt((A_copy[j,j]**2)+(A_copy[k,j]**2))
        # theta = np.arctan(A_copy[k,j]/A_copy[j,j])
        theta = np.arctan2(A_copy[k,j],A_copy[j,j])
        # A_copy[j,j] = d
        # A_copy[k,j] = 0
        Q_temp = np.eye(Shape[0])
        Q_temp[j,j] = np.cos(theta)
        Q_temp[k,k] = np.cos(theta)
        Q_temp[j,k] = -np.sin(theta)
        Q_temp[k,j] = np.sin(theta)
        Q = Q@Q_temp
        A_copy = Q_temp.T@A_copy
        # print(f"A: \n{A_copy}\n")
        # print(f"Q_temp: \n{Q_temp}\n")
        # print(f"j: {j}, k: {k}, d: {d}\n")
        # sns.heatmap(A_copy,cmap="viridis",annot=True,vmin=-1.5,vmax=1.5)
        sns.heatmap(A_copy,cmap="viridis",vmin=Min,vmax=Max)
        plt.savefig(f"Plot_{ctr}.png")
        Image_Files.append(f"Plot_{ctr}.png")
        ctr = ctr + 1
        plt.show()
```

```python
animation = ImageSequenceClip(Image_Files, fps = 10)
animation.write_videofile(f"Movie_Rotator_{Shape[0]}.mp4")
```
